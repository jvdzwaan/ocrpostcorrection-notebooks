````python
import tempfile

from ocrpostcorrection.icdar_data import extract_icdar_data, generate_data

with tempfile.TemporaryDirectory() as tmp_dir:
    data_path, test_path = extract_icdar_data(tmp_dir, '../data/raw/ICDAR2019_POCR_competition_dataset.zip')

    #data, _ = generate_data(data_path)
    data_test, _ = generate_data(test_path)
````

````python
from ocrpostcorrection.utils import extract_icdar_output

def create_icdar_output_all_tokens_incorrect(data):
    output = {}
    for key, text_obj in data.items():
        label_str = "".join(["1" for t in text_obj.input_tokens])
        output[key] = extract_icdar_output(label_str, data[key].input_tokens)
    return output
````

````python
test_output = create_icdar_output_all_tokens_incorrect(data_test)
````

````python
import json

out_file = 'data/results_all_input_tokens.json'

with open(out_file, 'w') as f:
    json.dump(test_output, f)
````
